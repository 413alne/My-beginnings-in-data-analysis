```python
import seaborn as sns
```

```python
data = sns.load_dataset('titanic')
```

```python
data
```

```python
data.tail(10)
```

```python
sns.set(style="darkgrid")
histogram = sns.histplot(x='class', data=data)
```

```python
sns.set(style="darkgrid")
plot = sns.relplot(x='age', y='fare', hue = 'sex', data=data)
```

```python
sns.set(style="darkgrid")
plot = sns.relplot(x='age', y='fare', col = 'sex', data=data)
```
