```python
columns_to_convert = ['DayOfWeek', 'Make', 'AccidentArea']

for column in columns_to_convert:
    df[column] = df[column].astype('category')

print(df.dtypes)
```

```python
df = df.rename(columns = {'DAY_OF_WEEK': 'DayOfWeek'})
df.head(3)
```

```python
df = df.rename(columns = {'make' : 'Make'})
df.head(3)
```

```python
df['make'] = df['make'].str.lower()
df.head(3)
```

```python
df['DayOfWeek'] = df['DayOfWeek'].str.upper()
df.head(3)
```

```python
frauds_df_honda = df[df['Make'].str.contains('hon', na=False)]
frauds_df_honda
```

```python
frauds_df_honda = frauds_df_honda.copy()
frauds_df_honda['PolicyRepNumber'] = frauds_df_honda['PolicyNumber'] + frauds_df_honda['RepNumber']
frauds_df_honda.head(6)
```
