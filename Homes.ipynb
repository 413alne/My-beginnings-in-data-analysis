```python
df.hp = pd.read_csv('home_prices.csv')
print(df.hp.head(5))
```

```python
df = pd.merge(df.hd, df.hp, left_on='id', right_on='house_id')
print(df.head(5))
```

```python
price_grade_df = df.groupby('grade')['price'].mean()
print(price_grade_df)
```
